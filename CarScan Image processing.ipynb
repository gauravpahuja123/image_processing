### Changing the background

```python
change_bg = alter_bg(model_type = "pb")
```

```python
change_bg.load_pascalvoc_model("xception_pascalvoc.pb")
```

```python
change_bg.change_bg_img(f_image_path = "///home/gauravpahuja/Documents/python/view1.jpeg",b_image_path = "///home/gauravpahuja/Documents/python/bg.png"
, output_image_name="view_no1_changed_bg.jpg")
```

```python
img1_changed_bg = io.imread("///home/gauravpahuja/Documents/python/view_no1_changed_bg.jpg")
```

```python
plt.imshow(img1_changed_bg)
```

### Removing the background

```python
im1 = Image.open(r'///home/gauravpahuja/Documents/python/view_no1_changed_bg.jpg')
im1.save(r'///home/gauravpahuja/Documents/python/new_view_no_1_png.png')
```

```python
def convertImg():
    img_1 = Image.open("///home/gauravpahuja/Documents/python/new_view_no_1_png.png")
    img_1 = img_1.convert("RGBA")
    
    data = img_1.getdata()
    
    newData = []
    
    for item in data:
        if item[0] == 255 and item[1] == 255 and item[2] == 255:
            newData.append((255,255,255,0))
        else:
            newData.append(item)
    img_1.putdata(newData)
    img_1.save("///home/gauravpahuja/Documents/python/new_view_no_1_removed_bg.png")
```

```python
convertImg()
```

```python
img1_removed_bg = io.imread("///home/gauravpahuja/Documents/python/new_view_no_1_removed_bg.png")
```

```python
plt.imshow(img1_removed_bg)
```
